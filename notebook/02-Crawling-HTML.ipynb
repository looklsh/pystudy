```python
# HTTP://media.daum.net에서 html받아와서 주요이슈 확인
from urllib.request import urlopen
from bs4 import BeautifulSoup

#http요청
req = urlopen("http://media.daum.net/")
#http 응답코드 확인 geetode()
print("HTTP Response Code:", req.getcode())

#만약 응답코드가 200이면 html받아오기
if req.getcode() == 200:
    #html 받아오기
   
    html = req.read()
     # encoding변환
    html = html.decode("utf-8")
    print(html)
else:
    print("HTTP ERROR")
```

```python
# 분석을 위해서 단순 html 텍스트를 -> DOM 트리 객체 형태로 전환
#BEAUTIFULsOUP
soup = BeautifulSoup(html, "html.parser") # html 파서로 텍스트를 객체로 전환
#print(soup)
#페이지 정보의 확인
print("페이지의 타이틀:", soup.title.text)

# 주요 이슈 selector: ul.list_issue
list_issue = soup.select("ul.list_issue")
#print(list_issue, type(list_issue))
titles = list_issue[0].select("a")
print("Count of titles", len(titles))

from bs4 import Tag

for title in titles:
    #print(type(title)) # a tag (include spans)
    for child in title.children:
        if isinstance(child, Tag):
            child.decompose() # 부모 태그에서 분리
    print("이슈 타이틀:", title.next.strip())
        #print(child, type(child))
    #print("이슈 타이틀", title.text.strip())
```

```python
# 대상 페이지: https://movie.naver.com/movie/running/current.nhn
# 최신 영화 리스트를 추출

# html text 받아오기
req = urlopen("https://movie.naver.com/movie/running/current.nhn")
html = req.read().decode("utf-8")

# 페이지의 일부 확인
#print(html.strip()[:100])

soup = BeautifulSoup(html, "html.parser")
# 페이지의 정보 확인
print("페이지 타이틀:", soup.title)
print("page title name:", soup.title.name) #태그의 이름
print("page title string:", soup.title.string)
print("특정 태그의 부모:", type(soup.title.parent))

# 페이지 소스를 예쁘게 출력하기: prettify()
print("네이버 영화:")
print(soup.prettify()[:200])
```

```python
# 분석과 추출: 영화랭킹
# movie_lst = soup.find("ul", attrs={"class": "lst_detail_t1"})
# print(type(movie_lst))

#titles = movie_lst.find("dt", attrs={"class": "tit"})
#print(type(titles))

#titles = soup.select("dt.tit")
#print(type(titles))

currents = soup.select("ul.lst_detail_t1")[0]
titles = currents.select("dt", attrs={"class": "tlt"})

movies = []

for title in titles:
    for child in title.children:
        if child.name == "a":
            #a tag를 찾음
            #print(child.text)
            movie_title = child.text
            movie_href = child['href']
            
            #print("movie url:", movie_href)
            movie = {"title": movie_title, "url": movie_href }
            movies.append(movie)

#영화 랭킹 출력
print("현재 영화 랭킹")
for index, movie in enumerate(movies):
    print("{}위 - {}".format(index + 1, movie))
            
```

```python
# 네이버 영화 페이지에서 포스터 이미지를 크롤링
print(html.strip()[:100])

# 저장 디렉토리 /crawl
import os

if not os.path.exists("/crawl"): # crawl 디렉터리가 없으면
    os.mkdir("/crawl")

#html -> 객체로 : 뷰티풀숲
soup = BeautifulSoup(html, "html.parser")
movies = soup.select("ul.lst_detail_t1 > li")

from urllib.request import urlretrieve

#print(movies[0])
# 개별 영화 loop
for movie in movies:
    img = movie.find("img")
    #print("IMG", img)
    img_name = img['src']
    img_alt = img['alt']
    #print(img_name)
    #print(img_alt)
    
    if len(img_name.split("?")) == 1:
        pass
    else: #이미지 네임에 ? 붙어있다
        img_name = img_name.split("?")[0]
    
    #print("이미지 파일명:", img_name)
    # 확장자를 추출
    ext = img_name.split(".")[-1]
    #print("이미지 확장자:", ext)
    #img의 alt의 값을 파일명으로 사용하기 적절하도록 변경
    import re
    #불용문자를 -로 교체
    fname = re.sub(r"[\s:,?]", "-", img_alt.strip())
    #최종 저장용 파일명을 생성
    full_name = "{}.{}".format(fname, ext)
    #print("최종 파일명:", full_name)
    
    full_name = "/".join(["/crawl", full_name])
    #print("최종 파일명:", full_name)
    
    #다운로드: urlretrieve(이미지url, 저장할 파일명)
    urlretrieve(img['src'], full_name)

print("=====이미지 저장완료")
        
```
